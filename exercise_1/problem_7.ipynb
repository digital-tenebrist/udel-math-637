## Imports and Globals

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

DATA_DIR='/data/udel-ms-data-science/math-637'
FILE_LIST=[f'exercise_6_data_{i}.txt' for i in range(1,5)]
```

### Read all data files into array

```python
DF_LIST=[pd.read_csv(f,header=None,delim_whitespace=True).transpose() for f in [f'{DATA_DIR}/{fn}' for fn in FILE_LIST]]
```

### Run PCA on each data set in turn

1. Index 0 : Intersecting Ellipses
1. Index 1 : Leading Edge Noisy Square Wave
1. Index 2 : IDK
1. Index 3 : Lemniscates - Infinity Symbols

### Index 0

```python
np.set_printoptions(precision=4,floatmode='fixed')

df=DF_LIST[0]
df=df-df.mean(axis=0)

pca = PCA(n_components=30)
pca.fit(df)

print(f'First eigenvector')
print(f'{pca.components_[0]}')

print(f'First eigenvalue          : {pca.singular_values_[0]:0.4f}')
print(f'Explained Variance Ratio  : {pca.explained_variance_ratio_[0]:0.4f}')
print(f'Sum of explained variance : {sum(pca.explained_variance_ratio_)}')

np.set_printoptions()

proj=df.dot(pca.components_[0])
proj.plot()
```
